## 4. Data Transformation

- Since the response variable has non linear relationship with all predictor variables, this has been log transformed before model building.

```python
data['log_median_house_value'] = [math.log(x) for x in data.median_house_value]
```

```python
#droping median house value
data = data.drop('median_house_value', axis = 1)
```

```python
data.head()
```

## 5. Splitting the data for training and testing

Since the primary goal is to build the most predictive model I partitioned the data set into a training set and a test set. The test set is used in comparing candidate models and the training set is used in training the selected models. Based on the performance of each model on the test set, the best model will be selected.  

Therefore, a partitioning strategy involving a 70% - 30% split was adopted, with 70% of the data allocated for  model training, while the remaining 30% was reserved for assessing each model's performance.

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(data.iloc[:,:8], data.log_median_house_value, random_state=123, shuffle=True, train_size=0.7, test_size=0.3)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

## 6. Training several regression models

During this phase, I trained the following predictive models: Multiple linear regression, K-Nearest Neighbors (KNN), Regression Trees, Random Forest and Boosting. The performance of each model was assessed on the test set and later selected the best model that yielded the least RMSE on the test set.

### 6.1 Multiple Linear Regression

Fitting the model

```python
mlr = LinearRegression()
mlr
```

```python
mlr.fit(X =Xtrain, y=ytrain )
```

The model coefficient estimates are summarized below:

```python
print(f"{'Coefficient':<19}{'Estimate'}\n")
print(f"{'Intercept':<19}{round(mlr.intercept_,3)}")
for i in range(len(mlr.feature_names_in_)):
    print(f"{mlr.feature_names_in_[i]:<19}{(mlr.coef_[i])}")
    
```

### 6.2 K-Nearest Neighbors (KNN)

To find the optimal K of the final model, the training data was first preprocessed standardizing every numerical variable. Then used 10-fold cross validation to obtain the optimal value of K through a grid search. Mean Square Error (MSE) was used to select the optimal model using the smallest value.

The values of K tried include: {3, 5, 7, 9, 15, 20, 40, 50}

```python
k_values = [3, 5, 7, 9, 15, 20, 40, 50]
```

Performing 10 fold cross validation to find the optimal K

```python
for k in  k_values:
    kfold = KFold(n_splits=10, shuffle=True, random_state=123)
    knn = KNeighborsRegressor(n_neighbors=k)
    score = cross_val_score(knn, X= Xtrain, y= ytrain, cv = kfold, scoring = 'neg_mean_squared_error')
    print(f"k = {k}: MSE = {round((-1*score).mean(),3)}")
```

K= 20 yields the minimum MSE therefore it's chosen as the optimal K. So I refit KNN with K = 20 as the final model

**Fitting the 20-NN**

```python
knn_final = KNeighborsRegressor(
    n_neighbors= 20
)
```

```python
knn_final.fit(X=Xtrain, y= ytrain)
```

### 6.3 Regression Trees

To obtain the optimal tree structure, and also avoid the overfitting of the regression trees on the test set, I used cost complexity pruning which penalizes Regression tree for having too many nodes by removing sections of the tree that provide little power to predict instances. Cost complexity pruning was implemented through 10-fold cross validation where MSE was used to select the optimal subtree model using the smallest value.

```python
rt = DecisionTreeRegressor(ccp_alpha=0.005417031)
```

```python
rt.fit(Xtrain, ytrain)
```

The resulting final model is visualized below

```python
ax = plt.subplots(figsize=(12,12))[1]
lt = plot_tree(rt,feature_names=data.columns[:8], ax=ax)
plt.show()
```

- According to the above regression tree, median income is the most important feature in predicting the median house value . 

### 6.4 Random Forest Regressor

Random Forest is an ensemble model built by generating B bootstrap samples from the training data set and then decision trees are fitted on each bootstrap sample such that each time a split in a tree is considered, the full set of p predictors is not considered. This ensures that the trees are decorrelated, hence reducing the variance. The final prediction is the average among the B models fitted on bootstrap samples.

This model was implemented using 10-fold cross-validation to determine the optimal tuning parameter max_features which controls the number of predictors to be tried at each split. MSE was used to select the optimal model using the smallest value.

```python
rf = RF(max_features=round(math.sqrt(8)), n_estimators=100, random_state=0)
```

```python
rf.fit(Xtrain, ytrain)
```

```python
**Feature importance**
```

This is a relative measure of the total decrease in node impurity that results from splits over that variable, averaged over all trees.

```python
feature_imp = pd.DataFrame({'importance':rf.feature_importances_}, index= data.columns[:8])
feature_imp.sort_values(by='importance', ascending=False)
```

- According to the above regression tree, median income is the most important feature in predicting the median house value , followed by latitude and longitude

### 6.5 Gradient Boosting Regressor

Gradient boosting combines many weak learners (such as shallow decision trees) to create a more accurate predictor.
It builds an ensemble of models in a stage-wise manner, optimizing an arbitrary differentiable loss function by iteratively adding new estimators that correct the errors of their predecessors.

```python
**Fitting the model**
```

```python
gb = GBR(n_estimators=100, learning_rate=0.001, max_depth=3,random_state=0)
```

```python
gb.fit(Xtrain, ytrain)
```

## 7. Evaluating the models' performance & selecting the best model

The model with smallest test RMSE was chosen as the best regression model. Below is the summary of each model's performance.

```python
models = {
    'Multiple Linear Regression': mlr,
    '20-NN': knn_final,
    'Regression Trees': rt,
    'Random Forest Regressor': rf,
    'Gradient Boosting Regressor': gb
}

print(f'{"Model":<28}  {"RMSE in dollars"}')
for object, model in models.items():
    predictions = model.predict(Xtest)
    predictions = [math.exp(x) for x in predictions]
    actual = [math.exp(x) for x in ytest]
    mse = mean_squared_error(predictions, actual)
    rmse = math.sqrt(mse)
    print(f'{object:<28}: ${round(rmse,3):>,}')
```

- Based on the above results, Random Forest Regressor yields the least RMSE value of $53,477.557. Therefore, it's the best model for predicting a new home's median house value in the future.
