# CLASSIFICATION: Spam Detection: A Machine Learning Approach for Email Classification

**About**

This problem is about classifying emails into spam or normal (or non-spam). There are 57 predictors (the first 57 columns) and the response variable (58th column) is binary denoting whether the e-mail was considered spam (1) or not (0). All predictors are quantitative and majority of them (48 out of 57) give the percentages of words in the email that match a specific key word such as “internet”, “address” or “business

**The goal is to classify an email as spam or non-spam.**

# 1.Importing the required libraries

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier as RF, GradientBoostingClassifier as GBR
```

## 4. Data Transformation

Each predictor has been transformed using the transformation function; ln(x + 0.1)

```python
predictors = data.columns[:57]
for p in predictors:
    data[p] = [math.log(x + 0.1) for x in data[p]]
```

```python
data.head()
```

## 5. Splitting the data for training and testing

Since the primary goal is to build the most predictive model I partitioned the data set into a training set and a test set. The test set is used in comparing candidate models and the training set is used in training the selected models. Based on the performance of each model on the test set, the best model will be selected.

Therefore, a partitioning strategy involving a 70% - 30% split was adopted, with 70% of the data allocated for model training, while the remaining 30% was reserved for assessing each model's performance.

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(data.iloc[:,:57], data.V58, random_state=123, shuffle=True, train_size=0.7, test_size=0.3)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

## 6. Training several classification models

During this phase, I trained the following predictive models: Logistic regression, Classification Trees, Random Forest and Gradient Boosting classifiers. The performance of each model was assessed on the test set and later selected the best model that yielded the highest accuracy on the test set.

### 6.1 Logistic Regression

```python
lgr = LogisticRegression(max_iter=1000)
lgr
```

```python
lgr.fit(Xtrain, ytrain)
```

The model coefficient estimates are summarized below:

```python
print(f"{'Coefficient':<14}{'Estimate':7}\n")
print(f"{'Intercept':<14}{round(lgr.intercept_[0], 4):>7}")
for idx, v in enumerate((lgr.feature_names_in_)):
    print(f"{v:<14}{round(lgr.coef_.ravel()[idx], 4):>7}")
```

### 6.2 Classification Trees

```python
ct = DecisionTreeClassifier(ccp_alpha= 0.005417)
ct
```

```python
ct.fit(Xtrain, ytrain)
```

The resulting final model is visualized below

```python
ax = plt.subplots(figsize=(12,12))[1]
lt = plot_tree(ct,feature_names=data.columns[:57], ax=ax)
plt.show()
```

- Variable V52 is the most important variable in detecting spam emails.

### 6.3 Random Forest Regressor

```python
rf = RF(max_features=round(math.sqrt(8)), n_estimators=100, random_state=0)
```

```python
rf.fit(Xtrain, ytrain)
```

**Feature importance**

This is a relative measure of the total decrease in node impurity that results from splits over that variable, averaged over all trees.

```python
feature_imp = pd.DataFrame({'importance':rf.feature_importances_}, index= data.columns[:57])
feature_imp.sort_values(by='importance', ascending=False).head()
```

- Variable V52 is the most important variable in detecting spam emails.

### 6.4 Gradient Boosting Classifier

Gradient boosting combines many weak learners (such as shallow decision trees) to create a more accurate predictor.
It builds an ensemble of models in a stage-wise manner, optimizing an arbitrary differentiable loss function by iteratively adding new estimators that correct the errors of their predecessors.

```python
gb = GBR(n_estimators=100, learning_rate=0.001, max_depth=3,random_state=0)
```

```python
gb.fit(Xtrain, ytrain)
```

## 7. Evaluating the models' performance & selecting the best model

The model with highest test accuracy was chosen as the best classification model. Below is the summary of each model's performance.

```python
models = {
    'Logistic Regression': lgr,
    'Classification Trees': ct,
    'Random Forest Classifier': rf,
    'Gradient Boosting Classifier': gb
}

print(f'{"Model":<28}   {"Accuracy (%)"}')
for object, model in models.items():
    predictions = model.predict(Xtest)
    acc = accuracy_score(predictions, ytest)
    print(f'{object:<29}: {acc:>11.2%}')
```

- Based on the above results, Random Forest Regressor yields the highest accuracy score of **94.90%**. Therefore, it's the best model for classifying email as spam or non-spam e in the future.
