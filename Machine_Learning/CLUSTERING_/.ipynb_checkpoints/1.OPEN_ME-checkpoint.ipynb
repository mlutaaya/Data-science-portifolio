# CLUSTERING: Cereal Groupings: Uncovering Breakfast Patterns

**About**

This problem is about grouping selected 43 break fast cereals using cluster analysis. The data set at hand contains measurements for 43 breakfast cereals.

***The goal is to identify the appropriate cereal groups***

## 1. Loading libraries

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
```

## 2. Data Importation and Transformation

This step involves importating the data, excluding non-required variables; `Brand`, `Manufacturer` and `Group` and standardizing all variables in preparation for the next steps.

```python
#loading data
cereal = pd.read_csv('cereal.csv')
cereal.head()
```

```python
#droping Brand, Manufacturer and Group variables.
cereal = cereal.drop(['Brand', 'Manufacturer', 'Group'],axis =1)
cereal.head()
```

Standardizing the remaining 8 predictors to have mean zero and standard deviation one

```python
pd.set_option('display.precision', 3)
for var in cereal.columns:
    cereal[var] = stats.zscore(cereal[var])
cereal.head()
```

## 3. Dimensionality Reduction with PCA

Principal Component Analysis (PCA) analyze a dataset’s features and reduce them to the specified number of
dimensions For this project I choose 2 principal components for visualization purposes. .

```python
 pca = PCA(n_components=2, random_state=123)
```

```python
pca_df = pca.fit_transform(cereal)
pca_df.shape
```

```python
pd.set_option('display.precision', 3)
pca_df = pd.DataFrame(pca_df, columns=['PC1', 'PC2'])
pca_df.head()
```

## 4. Identifying plausible Number Clusters Through Hierarchical Clustering

I considered Hierarchically clustering the cereals using the complete linkage and the Euclidean distance

```python
linkage_data = linkage(cereal, method='complete', metric='euclidean')
```

Displaying the resulting dendogram and deciding on choosing an appropriate height level at which to cut it

```python
dendrogram(linkage_data)
plt.hlines(y=7, xmin=0, xmax=10000, colors='red')
plt.show()
```

- Based on the above dendrogram, the appropriate height to cut the dendrogram is 7 because a horizontal line at height = 7, transverse the maximum distance vertically without intersecting a cluster. As a result, this suggests 3 possible clusters.

## 5. KMeans Clustering with K = 3

Since Hierarchical Clustering results suggest 3 clusters, a Kmeans clustering is implemented with K = 3 as shown below.

```python
kmeans = KMeans(n_clusters=3, random_state=123)
```

```python
kmeans.fit(cereal)
```

Finally, ploting the 2nd principal component scores versus the 1st, coloring the points based on the obtained KMeans cluster assignments.

```python
pca_df['Group'] = kmeans.labels_
```

```python
colors = ['#747FE3', '#8EE35D', '#E37346']
_ = sns.scatterplot(data = pca_df, x='PC2', y='PC1', hue='Group', legend='brief', palette=colors)
```

- Based on the plot above, group 0 dominates in terms of the number of cereal breakfast followed by group 2 and then group 1 is rare.
